```python
import pandas as pd
```

Read the list of propositions and contributions

```python
props = pd.read_csv("http://www.firstpythonnotebook.org/_static/committees.csv")
```

```python
props.head()
```

```python
contribs = pd.read_csv("file:///Users/elsengel/Desktop/contributions.csv")
```

```python
contribs.head()
```

Count the number of committees for every proposition

```python
props.prop_name.value_counts().reset_index()
```

Choose one proposition to focus on

```python
prop67 = props[props.prop_name == "PROPOSITION 067- REFERENDUM TO OVERTURN BAN ON SINGLE-USE PLASTIC BAGS."]
```

```python
prop67.head(10)
```

merge proposition 67 table with contributors table 

```python
merged = pd.merge(prop67, contribs, on="calaccess_committee_id")
```

```python
merged.head()
```

Now the magic starts!!

```python
merged.amount.sum()
```

```python
merged.committee_position.value_counts()
```

Look at differences between support and oppose donations

```python
support = merged[merged.committee_position == "SUPPORT"]
```

```python
oppose = merged[merged.committee_position == "OPPOSE"]
```

```python
support.amount.sum()
```

```python
oppose.amount.sum()
```

```python
support.amount.sum() / merged.amount.sum()
```

Make top 10

```python
merged.sort_values("amount", ascending=False).head(10)
```

```python
oppose.sort_values("amount", ascending=False).head(5)
```

Which committee raised the most money?

```python
merged.groupby("committee_name_x").amount.sum().reset_index().sort_values("amount", ascending=False)
```

Who were the biggest donors?

```python
merged.groupby(["contributor_lastname","committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(11)
```

```python
support.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(11)
```

```python
oppose.groupby(["contributor_lastname", "committee_position"]).amount.sum().reset_index().sort_values("amount", ascending=False).head(11)
```
